```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit, execute, Aer, IBMQ
from qiskit.tools.jupyter import *
from qiskit.visualization import *
```

```python
q = QuantumRegister(3) # 3 qubits
c = ClassicalRegister(3) # 3 bits
qc = QuantumCircuit(q,c)
qc.x(range(2))
qc.ccx(0,1,2)
qc.measure(range(3), range(3))
qc.draw(output = 'mpl') # draw circuit
```

```python
backend = Aer.get_backend('qasm_simulator')
job = execute(qc, backend, shots=1000000)
count = job.result().get_counts(qc)
plot_histogram(count)
```

```python
q = QuantumRegister(4) # 4 qubits
c = ClassicalRegister(3) # 3 bits
qc = QuantumCircuit(q,c)
qc.x(3) # เตรียม state |1>
qc.barrier() # กั้นเป็นส่วนๆ เพื่อความสวยงาม
qc.h(range(4)) # H gates
qc.x(2) # oracle, f(x)=1
qc.barrier() # กั้นเป็นส่วนๆ เพื่อความสวยงาม
qc.h(range(3)) # H gates
qc.measure(range(3), range(3))
qc.draw(output = 'mpl', plot_barriers=False)
```

```python
job = execute(qc, backend, shots=1000000)
count = job.result().get_counts(qc)
plot_histogram(count)
```
