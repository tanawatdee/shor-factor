```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit, execute, Aer, IBMQ
from qiskit.tools.jupyter import *
from qiskit.visualization import *
```

```python
q = QuantumRegister(1) # 1 qubit
c = ClassicalRegister(1) # 1 bit
qc = QuantumCircuit(q,c)
qc.draw(output = 'mpl') # draw circuit
```

```python
qc.x(0) # หรือ qc.x(q[0]) ก็ได้
qc.draw(output = 'mpl')
```

```python
qc.measure(0, 0) # หรือ qc.x(q[0], c[0]) ก็ได้
qc.draw(output = 'mpl')
```

```python
Aer.backends()
```

```python
backend = Aer.get_backend('qasm_simulator')
job = execute(qc, backend, shots=1000000)
count = job.result().get_counts(qc)
plot_histogram(count)
```

```python
qc = QuantumCircuit(q,c) # สร้างวงจรใหม่
qc.h(0) # หรือ qc.h(q[0]) ก็ได้
qc.measure(0, 0) # หรือ qc.x(q[0], c[0]) ก็ได้
qc.draw(output = 'mpl')
```

```python
backend = Aer.get_backend('qasm_simulator')
job = execute(qc, backend, shots=1000000)
count = job.result().get_counts(qc)
plot_histogram(count)
```
